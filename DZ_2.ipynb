```python
# Загружаем библиотеки:
# для работы с математикой
import numpy as np
```

```python
# Фиксируем начальное условие для генератора случайных чисел
np.random.seed(42)
```

```python
# Создайте случайную матрицу А1 размерностью 5 на 5
A1 = np.random.randint(0, 9, size = (5, 5))
print(f'A1 =\n{A1}')
# Определитель, ранг и собственные значения
print(f'det A1 = {np.linalg.det(A1)}')
print(f'rang A1 = {np.linalg.matrix_rank(A1)}')
print(f'Собственные значения: {np.linalg.eigvals(A1)}\nСобственные значения λ4 и λ5 комплексные')
```

```python
# Транспонируйте матрицу A1
A1_T = np.transpose(A1)
print(f'A1_T =\n{A1}')
# Определитель, ранг и собственные значения
print(f'det A1_T = {np.linalg.det(A1_T)}\nОпределители транспонированной и исходной матриц равны')
print(f'rang A1_T = {np.linalg.matrix_rank(A1_T)}\nРанги транспонированной и исходной матриц равны')
print(f'Собственные значения: {np.linalg.eigvals(A1_T)}\nСобственные значения транспонированной и исходной матриц равны, но в данном случае часть из них переставлены местами')
```

```python
# Найдите обратную матрицу для А1
A1_inv = np.linalg.inv(A1)
print(f'A1_inv =\n{A1_inv}')
# Определитель, ранг и собственные значения
print(f'det A1_inv = {np.linalg.det(A1_inv)}\nОпределитель обратной матрицы равен определителю исходной матрицы в степени -1: 1 / det A1 = {np.linalg.det(A1) ** -1}')
print(f'rang A1_inv = {np.linalg.matrix_rank(A1_inv)}')
print(f'Собственные значения: {np.linalg.eigvals(A1_inv)}')
print(f'Произведение исходной и обратной матриц даёт единичную матрицу:\nA1*A1_inv =\n{np.dot(A1, A1_inv)}')
```

```python
# Создайте диагональную матрицу - поэлементно перемножив единичную
# матрицу I и случайную матрицу A2 размерности 4 на 4
I = np.eye(4)
print(f'I =\n{I}')
A2 = np.random.randint(0, 9, size = (4, 4))
print(f'A2 =\n{A2}')
print(f'I*A2 =\n{I*A2}')
# Определитель, ранг и собственные значения
print(f'det I*A2 = {np.linalg.det(I*A2)}')
print(f'rang I*A2 = {np.linalg.matrix_rank(I*A2)}')
print(f'Собственные значения: {np.linalg.eigvals(I*A2)}')
```

```python
# Создайте нулевую матрицу размерности 3 на 3
Z = np.zeros((3, 3))
print(f'Z =\n{Z}')
# Определитель, ранг и собственные значения
print(f'det Z = {np.linalg.det(Z)}')
print(f'rang Z = {np.linalg.matrix_rank(Z)}')
print(f'Собственные значения: {np.linalg.eigvals(Z)}')
print('Определитель, ранг и собственные значения равны нулю!')
```
